```python
import re
import sys
import codecs
import nltk
from nltk.corpus import stopwords
import pandas as pd
```

```python
data = pd.read_csv('clean_sopwords_Index_words_eng.csv')
data.head()
```

```python
index = list(data.Index)[:5_00]
res = {}
```

```python
def update_dict(dictionary:dict, key:str, value:int) -> None:
    '''Обновление словаря'''
    value_in_dict = dictionary.get(key, 0)
    dictionary.update({key : value + value_in_dict})
```

```python
for i in index:
    with open(f"dataset_eng/{i}.txt", "r", newline="") as file:
        text = file.read()
        
        # удаление ссылок
        text = re.sub(r'(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:\'".,<>?«»""‘’]))', '', text)
        # удаление всех символов кроме букв
        text = re.sub(r"[^a-zA-Z]+", " ", text)
        
        words = nltk.word_tokenize(text)
        
        # удаление все слов длинной меньших 2 символов
        words = [word.lower() for word in words if len(word) > 1 and not word.isnumeric()]
        fdist = nltk.FreqDist(words)
        tmp_dict = dict(fdist.most_common(10))
        for key, value in  tmp_dict.items():
            update_dict(res, key, value)
```

```python
# сортировка элементов словаря от самого большего
sorted_frq_words = sorted(res.items(), key=lambda x: x[1])
sorted_frq_words.reverse()
sorted_frq_words
```

```python
stop_words = [i[0] for i in sorted_frq_words if i[1] > 5_00]
```

```python
with open('stop_words.txt', 'w') as fl:
    for item in stop_words:
        fl.write(f"{item} \n")
```
